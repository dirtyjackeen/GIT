# HELLO FRESH BENL VS BEFR

### *A clearer visualisation compairing HelloFresh BENL and HelloFresh BEFR*

```python
#packages
import pandas as pd
import csv
import matplotlib as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm
import numpy as np
import seaborn as sns
from statsmodels.formula.api import ols
import glob
import pycountry
from datetime import datetime

#chaning the number of columns that are displayed. get will tell you how much. 
pd.set_option("display.max_columns",50)
sns.set_style("darkgrid")

pd.options.mode.chained_assignment = None
```

```python
#organising cleaning
df1=pd.read_pickle("BENL.pkl")
df1["hf"]="BENL"
df1_total=df1.groupby(by=['month']).sum()
df1_total["Month"]=df1_total.index
df1_total["hf"]="BENL"
df1_total["margin_%"]=(df1_total["margin"]/df1_total["revenue"])*100
df1_total["margin_%"]=df1_total["margin_%"].round(2)


df2=pd.read_pickle("BEFR.pkl")
df2["hf"]="BEFR"
df2_total=df2.groupby(by=['month']).sum()
df2_total["Month"]=df2_total.index
df2_total["hf"]="BEFR"
df2_total["margin_%"]=(df2_total["margin"]/df2_total["revenue"])*100
df2_total["margin_%"]=df2_total["margin_%"].round(2)


df = pd.concat([df1,df2], ignore_index=True)
df_total = pd.concat([df1_total,df2_total], ignore_index=True)
```

```python
order=["15 Sep","15 Oct","15 Nov","15 Dec",
       "16 Jan","16 Feb","16 Mar","16 Apr","16 May","16 Jun","16 Jul","16 Aug","16 Sep","16 Oct","16 Nov","16 Dec",
       "17 Jan","17 Feb","17 Mar","17 Apr","17 May","17 Jun","17 Jul","17 Aug","17 Sep","17 Oct","17 Nov","17 Dec",
       "18 Jan","18 Feb","18 Mar","18 Apr","18 May","18 Jun"]
```

# Average Margin % per Campaign each Month

```python
plt.pyplot.figure(figsize=(16,6))
ha=sns.pointplot(x="month", y="marginP", hue="hf", palette="seismic", errwidth=.5, data=df, order=order)
ha.set_xticklabels(ha.get_xticklabels(), rotation=40, ha="right")
ha
```

# Total Margin % each Month

```python
plt.pyplot.figure(figsize=(16,6))
ha=sns.pointplot(x="Month", y="margin_%", hue="hf", palette="seismic", errwidth=.5, data=df_total, order=order)
ha.set_xticklabels(ha.get_xticklabels(), rotation=40, ha="right")
ha
```
